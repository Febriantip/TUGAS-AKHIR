```python
tokenized_tweet = df['full_text'].apply(lambda x: x.lower().split()) #tokenize

import gensim
model_w2v = gensim.models.Word2Vec(
            tokenized_tweet,
            vector_size=200,  # Change 'size' to 'vector_size'
            window=5,  # ukuran jendela
            min_count=1, #Lowering min_count to 1, so single occurrence words are considered
            sg=1,  # skip gram model
            hs=0,
            negative=10,  # jumlah contoh kata negatif
            workers=2,
            seed=34)

model_w2v.train(tokenized_tweet, total_examples=len(df['full_text']), epochs=20)

# Calling most_similar with the lowercase version
model_w2v.wv.most_similar(positive="obama")
```

```python
def word_vector(tokens, size):
    vec = np.zeros(size).reshape((1, size))
    count = 0.
    for word in tokens:
        try:
            vec += model_w2v[word].reshape((1, size))
            count += 1.
        except KeyError: # apabila ada kasus dimana token tidak ada dalam kosa kata

            continue
    if count != 0:
        vec /= count
    return vec
```

```python
def word_vector(tokens, size):
    vec = np.zeros(size).reshape((1, size))
    count = 0.
    for word in tokens:
        try:
            # Access the word vector using model_w2v.wv[word]
            vec += model_w2v.wv[word].reshape((1, size))
            count += 1.
        except KeyError: # apabila ada kasus dimana token tidak ada dalam kosa kata
            continue
    if count != 0:
        vec /= count
    return vec
```
